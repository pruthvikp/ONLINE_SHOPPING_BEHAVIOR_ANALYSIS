# Plots for the Dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
filename = 'activity_data.csv'
data = pd.read_csv(filename)
```

# Histograms for all numerical features

```python
data.hist(figsize=(14, 10), bins=30, edgecolor='k')
plt.tight_layout()
plt.show()
```

# Heatmap

```python
data = pd.read_csv(filename)

# Convert categorical columns to numerical values using one-hot encoding
data = pd.get_dummies(data, columns=["VisitorType", "Weekend"])

# Convert 'Month' to numerical values
month_index = {
    "Jan": 0, "Feb": 1, "Mar": 2, "Apr": 3, "May": 4, "June": 5,
    "Jul": 6, "Aug": 7, "Sep": 8, "Oct": 9, "Nov": 10, "Dec": 11
}
data['Month'] = data['Month'].map(month_index)

# Calculate the correlation matrix
corr_matrix = data.corr()

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, fmt='.2f')
plt.title("Correlation Matrix")
plt.show()
```

# Count plots for different features of the dataset

```python
data = pd.read_csv(filename)

# Convert 'Month' to numerical values
month_index = {
    "Jan": 0, "Feb": 1, "Mar": 2, "Apr": 3, "May": 4, "June": 5,
    "Jul": 6, "Aug": 7, "Sep": 8, "Oct": 9, "Nov": 10, "Dec": 11
}
data['Month'] = data['Month'].map(month_index)

# Correct the column names used in count plots
plt.figure(figsize=(14, 10))

plt.subplot(2, 2, 1)
sns.countplot(data=data, x='VisitorType', hue='Revenue', palette='Set2')
plt.title('Visitor Type Count')

plt.subplot(2, 2, 2)
sns.countplot(data=data, x='Weekend', hue='Revenue', palette='Set2')
plt.title('Weekend Visit Count')

plt.subplot(2, 2, 3)
sns.countplot(data=data, x='Month', hue='Revenue', palette='Set2')
plt.title('Month Visit Count')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

# Boxplots for numerical features to detect outliers

```python
plt.figure(figsize=(14, 10))

plt.subplot(2, 2, 1)
sns.boxplot(data=data, x='Revenue', y='BounceRates', palette='Set2')
plt.title('BounceRates Boxplot')

plt.subplot(2, 2, 2)
sns.boxplot(data=data, x='Revenue', y='ExitRates', palette='Set2')
plt.title('ExitRates Boxplot')

plt.subplot(2, 2, 3)
sns.boxplot(data=data, x='Revenue', y='PageValues', palette='Set2')
plt.title('PageValues Boxplot')

plt.subplot(2, 2, 4)
sns.boxplot(data=data, x='Revenue', y='SpecialDay', palette='Set2')
plt.title('SpecialDay Boxplot')

plt.tight_layout()
plt.show()
```
